```python
#get list of files in 'Methods' folder
files = os.listdir('Methods')
sentences = pd.DataFrame()
for file in files:
    #open file
    with open(f'Methods/{file}', 'r') as f:
        #get corpusid
        corpusid = file.split('_')[1].split('.')[0]
        #read file
        text = f.read()
        segments = text.split('\n')
        #tokenize
        for segment in segments:
            tokens = nltk.sent_tokenize(segment)
            #create dataframe
            df = pd.DataFrame(tokens, columns=['sentence'])
            df['corpusid'] = corpusid
            sentences = pd.concat([sentences, df])
```

```python
sentences.to_csv('sentences.csv')
```

```python
def has_proper_noun(sentence):
    words = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(words)
    for word, pos in pos_tags:
        if pos == 'NNP' or pos == 'NNPS':
            return True
    return False
```

```python
sentences['proper_noun'] = sentences['sentence'].apply(has_proper_noun)
```

```python
sentences[sentences['proper_noun']].to_csv('sentences_proper_noun.csv')
```
